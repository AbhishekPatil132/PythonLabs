```python
import numpy as np
```

```python
#taking input array
arr = np.arange(10)
print(arr)
```

```python
#finding the mean of given array
mean = np.mean(arr)
print("Mean:", mean)
```

```python
#finding the standard deviation of given array
std = np.std(arr)
print("Standard deviation:", std)
```

```python
#finding the variance of given array
var = np.var(arr)
print("Variance:", var)
```

**Mean, Standard deviation and Variance with using standard libraries**

```python
arr = np.arange(10)
print(arr)
```

```python
r1 = np.average(arr)
print("Mean:", r1)
```

```python
r2 = np.sqrt(np.average((arr - r1)**2))
print("Standard deviation:", r2)
```

```python
r3 = np.average((arr - r1)**2)
print("Variance:", r3)
```

**Normal distribution**

```python
import matplotlib.pyplot as plt
```

```python
#Creating a Function.
def normal_dist(arr , mean , std):
   prob_density=(np.pi*std) *np.exp(-0.5*((arr-mean)/std)**2)
   return prob_density


#Apply function to the data.
pdf =normal_dist(arr,mean,std)
x = arr
y = mean

#Plotting the Results
plt.plot(arr, pdf, color ='red')
plt.plot(y)
plt.xlabel('Data points')
plt.ylabel('Probability Density')
```

```python
# import required libraries
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

# Creating the distribution
data =np.arange(1,10,0.01)
pdf =norm.pdf(data , loc=5.3, scale =1)

#Visualizing the distribution

sb.set_style('darkgrid')
sb.lineplot(pdf, color='black')
plt.xlabel('Heights')

plt.ylabel('Probability Density')
```

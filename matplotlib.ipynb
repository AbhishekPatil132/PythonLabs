```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df.head()
```

```python
df.shape
```

**Age vs Income**

```python
# plotting the graph using 'plot'
x = df["housing_median_age"]
y = df["median_income"]
plt.plot(x,y)
plt.title("Age vs Income graph")
plt.xlabel("Median Age")
plt.ylabel("Median Income")
plt.show
```

```python
# plotting the graph using 'bar'
x = df["housing_median_age"]
y = df["median_income"]
plt.bar(x,y)
plt.title("Age vs Income graph")
plt.xlabel("Median Age")
plt.ylabel("Median Income")
plt.show
```

```python
plt.barh(x,y)
```

```python
# plotting the graph using 'scatter'
x = df["housing_median_age"]
y = df["median_income"]
plt.scatter(x,y)
plt.title("Age vs Income graph")
plt.xlabel("Median Age")
plt.ylabel("Median Income")
plt.show
```

```python
# plotting the graph using 'stem'
x = df["housing_median_age"]
y = df["median_income"]
plt.stem(x,y)
plt.title("Age vs Income graph")
plt.xlabel("Median Age")
plt.ylabel("Median Income")
plt.show
```

```python
# plotting the graph using 'plot'
x = df["median_income"]
y = df["households"]
plt.plot(x,y)
plt.title("Income vs Househlds graph")
plt.xlabel("Median Income")
plt.ylabel("Households")
plt.show
```

**Other basic graphs**

```python
#implementing the pie graph
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
```

```python
#using 'explode' parameter in pie
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explodes = (0.2, 0, 0, 0)
plt.pie(sizes, labels=labels, autopct="%1.1f%%", explode= explodes)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("car_crashes")
df.head()
```

```python
df.shape
```

**1. Distribution plot**

```python
sns.histplot(df["not_distracted"], kde=False)
```

**2. Joint plot**

```python
sns.jointplot(x='speeding', y='alcohol', data=df, kind='hex')
```

**3. KDE plot**

```python
sns.kdeplot(df['speeding'])
```

**4. Pair plot**

```python
sns.pairplot(df)
```

```python
tdf = sns.load_dataset('tips')
tdf
```

```python
sns.pairplot(tdf, hue="size", palette='Blues')
```

**Styling**

```python
sns.set_style('ticks')# white, darkgrid, whitegrid, ticks
plt.figure(figsize=(10,4))
sns.set_context(font_scale = 1)#set font size

sns.jointplot(x='speeding', y='alcohol', data= df, kind='reg')

#To remove X,Y axis from the graph
#sns.despine(left=True, bottom=True)
```

**Categorial plots**

**5. Bar plots**

```python
sns.barplot(x='sex', y='total_bill', data= tdf, estimator= np.std)
```

**6. Box plot**

```python
sns.boxplot(x='day', y='total_bill', data=tdf, hue='sex')
plt.legend(loc=1)
```

**7. violin plot**

```python
sns.violinplot(x='day', y='total_bill', data=tdf, hue='sex')
```

**8. Strip plot**

```python
plt.figure(figsize=(8,4))
sns.stripplot(x='day', y='total_bill', data=tdf, hue='sex', jitter=True, dodge=False)
#dodge
```

```python
plt.figure(figsize=(8,4))
sns.stripplot(x='day', y='total_bill', data=tdf, hue='sex', jitter=True, dodge=True)
#dodge
```

**9. Swarm plot**

```python
sns.swarmplot(x="day", y="total_bill", data=tdf, palette='magma')
```

**Matrix plots**

**10. Heat map**

```python
#heatmap requires the data in matrix format
crash_matrix = df.corr()
crash_matrix
```

```python
sns.heatmap(crash_matrix, annot=True, cmap="Blues")
```

```python
# creating heatmap using the pivot table
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights, cmap='Blues')
```
